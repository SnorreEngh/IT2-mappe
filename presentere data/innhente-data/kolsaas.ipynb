```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[1][0] # -> gir meg den f√∏rste tiden som er null 0 
```

```python
tid = []
h√∏yde = []

for i in range(0,len(tur)):
    tur2 = tur[i][0]
    tid.append(tur2)

for i in range(0, len(tur)):
    tur3 = tur[i][3]
    h√∏yde.append(tur3)

plt.plot(tid, h√∏yde)
plt.show()
```

```python
# x = []
# y = []
lengde = []
bredde = []

for linje in tur[1:]: # klammeparantesen betyr at vi skal hoppe over linje 1 i llisten 
    # h√∏yde = linje[3]
    # tid = linje[0]
    lengde.append(float(linje[2]))
    bredde.append(float(linje[1]))
    # x.append(tid)
    # y.append(h√∏yde)

plt.plot(lengde, bredde)
plt.show()
```
